```python
import matplotlib.pyplot as plt
import pandas as pd

from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Flatten,  MaxPooling2D, Conv2D

import tensorflow as tf

import numpy as np

from sklearn.model_selection import GridSearchCV
from keras.wrappers.scikit_learn import KerasClassifier
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if len(physical_devices) > 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

## Loading data

```python
# path to train dataset
train_path = 'digit-recognizer/train.csv'
# path to test dataset
test_path = 'digit-recognizer/test.csv'

# Read a comma-separated values (csv) file into pandas DataFrame
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# shape of tha data
print('Train shape: ', train_data.shape)
print('Test shape: ', test_data.shape)
train_data.info()
```

```python
if 'label' in train_data.columns:
    train_label = train_data['label']
    del train_data['label']
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
for i in range(9):
    data = train_data.iloc[i].to_numpy().reshape((28,28))
    axs[i//3, i%3].imshow(data, cmap='gray', vmin=0, vmax=255)
    axs[i//3, i%3].set_title(train_label.iloc[i], fontsize=20)
```

```python
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(train_data, train_label)

X_train = X_ros.to_numpy().reshape(X_ros.shape[0],28,28,1).astype('float32')
X_test = test_data.to_numpy().reshape(test_data.shape[0],28,28,1).astype('float32')

X_train /= 255
X_test /= 255

n_classes = 10
Y_train = keras.utils.to_categorical(y_ros.to_numpy(), n_classes)
```

```python
def create_model():
    model = Sequential()
    model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(28,28,1)) )
    model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
    model.add(MaxPooling2D(pool_size=(2,2)))
    model.add(Dropout(0.25))
    model.add(Flatten())          
    model.add(Dense(128, activation='relu'))
    model.add(Dropout(0.5))
    model.add(Dense(n_classes, activation='softmax'))
    model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
    return model
```

```python
model = KerasClassifier(build_fn=create_model, verbose=0)
```

```python
batch_size = [16, 32, 64, 128]
epochs = [10, 20, 40]
param_grid = dict(batch_size=batch_size, epochs=epochs)
grid = GridSearchCV(estimator=model, param_grid=param_grid, n_jobs=None, cv=3, verbose=1)
grid_result = grid.fit(X_train, Y_train)
```

```python
print("Best: %f using %s" % (grid_result.best_score_, grid_result.best_params_))
```

```python
model.fit(X_train, Y_train, 
          batch_size=grid_result.best_params_['batch_size'], 
          epochs=grid_result.best_params_['epochs'], 
          verbose=1)
```

```python
Y_test = model.predict(X_test)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
for i in range(9):
    data = test_data.iloc[i].to_numpy().reshape((28,28))
    axs[i//3, i%3].imshow(data, cmap='gray', vmin=0, vmax=255)
    axs[i//3, i%3].set_title(Y_test[i], fontsize=20)
```

```python
output = pd.DataFrame({'ImageId': (i for i in range(1,test_data.shape[0]+1)), 'Label': Y_test})
output.to_csv('submission_Grid_batch=[16,32,64,128]_epochs=[10,20,40].csv', index=False)
```
