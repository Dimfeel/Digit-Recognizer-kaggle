```python
import matplotlib.pyplot as plt
import pandas as pd

from imblearn.over_sampling import RandomOverSampler
from sklearn.model_selection import train_test_split

import keras
from keras.models import Sequential
from keras.layers import Dense, Dropout
from keras.layers import Flatten,  MaxPooling2D, Conv2D

import tensorflow as tf

import numpy as np
import cv2
```

```python
physical_devices = tf.config.experimental.list_physical_devices('GPU')
if len(physical_devices) > 0:
    tf.config.experimental.set_memory_growth(physical_devices[0], True)
```

## Loading data

```python
# path to train dataset
train_path = 'digit-recognizer/train.csv'
# path to test dataset
test_path = 'digit-recognizer/test.csv'

# Read a comma-separated values (csv) file into pandas DataFrame
train_data = pd.read_csv(train_path)
test_data = pd.read_csv(test_path)

# shape of tha data
print('Train shape: ', train_data.shape)
print('Test shape: ', test_data.shape)
train_data.info()
```

```python
if 'label' in train_data.columns:
    train_label = train_data['label']
    del train_data['label']
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
for i in range(9):
    data = train_data.iloc[i].to_numpy().reshape((28,28))
    axs[i//3, i%3].imshow(data, cmap='gray', vmin=0, vmax=255)
    axs[i//3, i%3].set_title(train_label.iloc[i], fontsize=20)
```

```python
Scale_value = 3
train_data_scale = np.zeros((train_data.shape[0], Scale_value*28*Scale_value*28))
test_data_scale = np.zeros((test_data.shape[0], Scale_value*28*Scale_value*28))

for i in range(train_data.shape[0]):
    data = cv2.resize(
        train_data.iloc[i].to_numpy().reshape((28,28)).astype('float32'), 
        dsize=(Scale_value*28, Scale_value*28), 
        interpolation=cv2.INTER_CUBIC)
    data = data.reshape((Scale_value*28*Scale_value*28))
    train_data_scale[i] = data
    
for i in range(test_data.shape[0]):
    data = cv2.resize(
        test_data.iloc[i].to_numpy().reshape((28,28)).astype('float32'), 
        dsize=(Scale_value*28, Scale_value*28), 
        interpolation=cv2.INTER_CUBIC)
    data = data.reshape((Scale_value*28*Scale_value*28))
    test_data_scale[i] = data
    
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
for i in range(9):
    data = train_data_scale[i].reshape(Scale_value*28, Scale_value*28)
    axs[i//3, i%3].imshow(data, cmap='gray', vmin=0, vmax=255)
    axs[i//3, i%3].set_title(train_label.iloc[i], fontsize=20)
```

```python
def cut_block(images, percent_size_min, percent_size_max):
    images_reshape = images.reshape(images.shape[0],Scale_value*28,Scale_value*28).astype('float32')
    cutoff_size = tf.random.uniform(shape = [images.shape[0], 2],
                                    minval = np.floor(percent_size_min*Scale_value*28), 
                                    maxval = np.floor(percent_size_max*Scale_value*28),
                                    dtype = tf.dtypes.int32).numpy()
    cutoff_pos_start = tf.random.uniform(shape = [images.shape[0], 2],
                                         minval = 0, 
                                         maxval = Scale_value*28,
                                         dtype = tf.dtypes.int32).numpy()
    cutoff_pos_end = np.add(cutoff_pos_start, cutoff_size)
    cut_images = images_reshape
    for i in range(images.shape[0]):
        cut_images[i, cutoff_pos_start[i,0]:cutoff_pos_end[i,0],cutoff_pos_start[i,1]:cutoff_pos_end[i,1]] = np.float32(0)
    cut_images = cut_images.reshape(images.shape[0], Scale_value*28*Scale_value*28)
    return cut_images
```

```python
def rotate_image(image, angle):
    image = image.reshape(Scale_value*28,Scale_value*28).astype('float32')
    image_center = tuple(np.array(image.shape[1::-1]) / 2)
    rot_mat = cv2.getRotationMatrix2D(image_center, angle, 1.0)
    result = cv2.warpAffine(image,
                            rot_mat,
                            image.shape[1::-1],
                            flags=cv2.INTER_LINEAR).reshape(Scale_value*28*Scale_value*28)
    return result
```

```python
def rotate_images(images, min_angle, max_angle):
    angles = tf.random.uniform(shape = [images.shape[0]],
                               minval = min_angle, 
                               maxval = max_angle,
                               dtype = tf.dtypes.int32).numpy()
    for i in range(images.shape[0]):
        images[i] = rotate_image(images[i],angles[i])
    return images
```

```python
def add_train_data_cut(train_data, train_label, percent_add):
    init_shape = train_data.shape[0]
    elem = tf.random.uniform(shape = [np.floor(init_shape*percent_add).astype('int32')],
                             minval = 0, 
                             maxval = train_data.shape[0]-1,
                             dtype = tf.dtypes.int32).numpy()
    data = cut_block(np.take(train_data, elem, 0), 0.3, 0.4)
    train_data = np.append(train_data, data, axis=0)
    train_label = np.append(train_label, train_label[elem], axis=0)
    print("Init train data size = ", init_shape)
    print("Final size = ", train_data.shape[0])
    print("Added ", np.floor(init_shape*percent_add).astype('int32'), " elements")
    return train_data, train_label
```

```python
def add_train_data_rotate(train_data, train_label, percent_add):
    init_shape = train_data.shape[0]
    elem = tf.random.uniform(shape = [np.floor(init_shape*percent_add).astype('int32')],
                             minval = 0, 
                             maxval = train_data.shape[0]-1,
                             dtype = tf.dtypes.int32).numpy()
    data = rotate_images(np.take(train_data, elem, 0), -30, 30)
    train_data = np.append(train_data, data, axis=0)
    train_label = np.append(train_label, train_label[elem], axis=0)
    print("Init train data size = ", init_shape)
    print("Final size = ", train_data.shape[0])
    print("Added ", np.floor(init_shape*percent_add).astype('int32'), " elements")
    return train_data, train_label
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,15))
elem = tf.random.uniform(shape = [3],
                         minval = 0, 
                         maxval = train_data_scale.shape[0]-1,
                         dtype = tf.dtypes.int32).numpy()
data = cut_block(np.take(train_data_scale, elem, 0), 0.3, 0.4).reshape(3, Scale_value*28,Scale_value*28)
for i in range(3):
    axs[i, 0].imshow(train_data_scale[elem[i]].reshape(Scale_value*28,Scale_value*28), cmap='gray', vmin=0, vmax=255)
    axs[i, 1].imshow(data[i], cmap='gray', vmin=0, vmax=255)
    axs[i, 0].set_title(train_label[elem[i]], fontsize=20)
    axs[i, 1].set_title(train_label[elem[i]], fontsize=20)
```

```python
fig, axs = plt.subplots(3, 2, figsize=(15,15))
elem = tf.random.uniform(shape = [3],
                         minval = 0, 
                         maxval = train_data_scale.shape[0]-1,
                         dtype = tf.dtypes.int32).numpy()
data = rotate_images(np.take(train_data_scale, elem, 0), -30, 30)
for i in range(3):
    axs[i, 0].imshow(train_data_scale[elem[i]].reshape(Scale_value*28,Scale_value*28), cmap='gray', vmin=0, vmax=255)
    axs[i, 1].imshow(data[i].reshape(Scale_value*28,Scale_value*28), cmap='gray', vmin=0, vmax=255)
    axs[i, 0].set_title(train_label[elem[i]], fontsize=20)
    axs[i, 1].set_title(train_label[elem[i]], fontsize=20)
```

```python
#train_data_scale, train_label = add_train_data_cut(train_data_scale, train_label, 0.15)
```

```python
train_data_scale, train_label = add_train_data_rotate(train_data_scale, train_label, 0.25)
```

```python
ros = RandomOverSampler(random_state=42)
X_ros, y_ros = ros.fit_resample(train_data_scale, train_label)

x_train, x_valid, y_train, y_valid = train_test_split(X_ros, y_ros, test_size=0.1, random_state=10)

X_train = x_train.reshape(x_train.shape[0],Scale_value*28,Scale_value*28,1).astype('float32')
X_valid = x_valid.reshape(x_valid.shape[0],Scale_value*28,Scale_value*28,1).astype('float32')
X_test = test_data_scale.reshape(test_data_scale.shape[0],Scale_value*28,Scale_value*28,1).astype('float32')


X_train /= 255
X_valid /= 255
X_test /= 255

n_classes = 10
Y_train = keras.utils.to_categorical(y_train, n_classes)
Y_valid = keras.utils.to_categorical(y_valid, n_classes)

del X_ros
del y_ros
del train_data_scale
del train_label
```

```python
model = Sequential()
model.add(Conv2D(32, kernel_size=(3,3), activation='relu', input_shape=(Scale_value*28,Scale_value*28,1)) )
model.add(Conv2D(64, kernel_size=(3,3), activation='relu'))
model.add(MaxPooling2D(pool_size=(2,2)))
model.add(Dropout(0.25))
model.add(Flatten())          
model.add(Dense(128, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(n_classes, activation='softmax'))
```

```python
model.compile(loss='categorical_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
model.fit(X_train, Y_train, batch_size=16, epochs=15, verbose=1, validation_data=(X_valid,Y_valid))
```

```python
Y_test = model.predict_classes(X_test)
```

```python
fig, axs = plt.subplots(3, 3, figsize=(15,15))
for i in range(9):
    data = test_data_scale[i].reshape((Scale_value*28,Scale_value*28))
    axs[i//3, i%3].imshow(data, cmap='gray', vmin=0, vmax=255)
    axs[i//3, i%3].set_title(Y_test[i], fontsize=20)
```
